```python
def timestampsDraw(column,number):
    plt.subplot(2,3,number+1)
    plt.plot(df['ActivityDate'], df[column],color = 'indigo', alpha = 0.6)
    plt.xticks(df['ActivityDate'],rotation = 90, fontsize = 7 )
    plt.yticks(rotation = 45, fontsize = 7 ) 
    plt.grid(axis = 'x', alpha = 0.2)
    plt.ylim(df[column].min()*0.3,df[column].max()*1.2)
    title = column + " Timestamps"
    plt.title(title,fontsize = 12)

# Make function to run timestampsDraw
def timestampsDrawList(list):
    plt.figure(figsize = (20,8))
    plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)
    for i in range(6):
        timestampsDraw(list[i],i)
    plt.show()

# Run function
columnsTimestamps = ['TotalSteps','TotalDistance','Calories','TotalMinutesAsleep','WeightKg','BMI']
timestampsDrawList(columnsTimestamps)
```

# Active Distance Active Minutes Percentage

```python
distanceVA = round(df['VeryActiveDistance'].mean(),2)
distanceMA = round(df['ModeratelyActiveDistance'].mean(),2)
distanceLA = round(df['LightActiveDistance'].mean(),2)
distanceSA = round(df['SedentaryActiveDistance'].mean(),2)
columnAD = [distanceVA,distanceMA,distanceLA,distanceSA]
labelAD = ['Very Active','Moderate Active','Light Active','Sedentary Active']

# Active Distance Data
minutesVA = round(df['VeryActiveMinutes'].mean(),2)
minutesFA = round(df['FairlyActiveMinutes'].mean(),2)
minutesLA = round(df['LightlyActiveMinutes'].mean(),2)
minutesS = round(df['SedentaryMinutes'].mean(),2)
columnAM = [minutesVA,minutesFA,minutesLA,minutesS]
labeAM = ['Very Active','Fairly Active','Lightly Active','Sedentary']

# Function to draw pie
def drawPie(data, text, label):
    plt.figure(figsize=(10,10))
    plt.pie(x = data, autopct = '%1.1f%%',pctdistance = 1.1,
            wedgeprops = {'linewidth':1.0, 'edgecolor':'black'},
            textprops  = {'color' : 'black', 'fontsize':10})
    legendTitle = text
    plt.legend(labels = label, title = legendTitle, loc = 'lower right')
    graphTitle = "Percentage of " + text 
    plt.title(graphTitle, fontsize = 30, y = 0.96)
    plt.show()
    
# Run function active distance
drawPie(columnAD,'Active Distance',labelAD)
```

```python
# Run function active distance
drawPie(columnAM,'Active Minutes',labeAM)
```

# Active Distance and Active Minutes

```python
def drawBoxplot(column,number):
    plt.subplot(2,2,number+1)
    sns.boxplot(data = df,y = df[column], color = 'indigo')

# Function to run drawBoxplot
def runDistribution(list):
    plt.figure(figsize = (15,8))
    for i in range(len(list)):
        drawBoxplot(list[i],i)
    

# Run function active distance
listColumnAM = ['VeryActiveDistance','ModeratelyActiveDistance','LightActiveDistance','SedentaryActiveDistance']
runDistribution(listColumnAM)
```

```python
#Active minute
listColumnAM = ['VeryActiveMinutes','FairlyActiveMinutes','LightlyActiveMinutes','SedentaryMinutes']
runDistribution(listColumnAM)
```

# Relationship between Calorie,Weight,BMI and Total Minutes Sleep with Total Steps

```python
#draw scatter
def drawScatter(var1,var2,number):
    plt.subplot(2,2,number+1)
    plt.scatter(data = df, x = var1, y = var2, color = 'indigo', alpha = 0.8, marker = "X", s = 70)
    plt.ylabel(var2)
    plt.xlabel(var1)

# Columns
listColumn = ['Calories','WeightKg','BMI','TotalMinutesAsleep']

# Run function
plt.figure(figsize = (16,8))
for i in range(4):
    drawScatter(listColumn[i],'TotalSteps',i)

plt.show()
```
